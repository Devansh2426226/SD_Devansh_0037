```python
class BudgetTracker:

    def __init__(self):
        self.users = {}

    # Register Users
    def register(self):
        n = int(input("How many users do you want to add? "))

        for i in range(n):
            name = input("Enter username: ")
            password = int(input("Enter password: "))
            self.users[name] = password

        print("\nUsers registered successfully!")

    # Login
    def login(self):
        print("\n---------- LOGIN ----------")
        username = input("Enter username: ")

        if username in self.users:
            password = int(input("Enter password: "))

            if self.users[username] == password:
                print("\nLogin Successful!")
                self.calculate_budget()
            else:
                print("Incorrect Password!")

        else:
            print("Username not found!")

    # Budget Calculation
    def calculate_budget(self):

        budget = float(input("\nEnter your total budget: ₹"))

        clothing = float(input("Enter Clothing Expense: ₹"))
        food = float(input("Enter Food Expense: ₹"))
        travel = float(input("Enter Travel Expense: ₹"))

        total_expense = clothing + food + travel
        remaining = budget - total_expense

        print("\n========== EXPENSE SUMMARY ==========")
        print(f"Budget            : ₹{budget}")
        print(f"Clothing Expense  : ₹{clothing}")
        print(f"Food Expense      : ₹{food}")
        print(f"Travel Expense    : ₹{travel}")
        print("--------------------------------------")
        print(f"Total Expense     : ₹{total_expense}")

        if total_expense > budget:
            print(f"\nYou exceeded your budget by ₹{total_expense-budget:.2f}")

        elif total_expense == budget:
            print("\nYou have used your complete budget.")

        else:
            percentage = (remaining / budget) * 100
            print(f"\nRemaining Budget  : ₹{remaining:.2f}")
            print(f"Budget Left       : {percentage:.2f}%")

        print("\nBudget calculation successful!")


# -------- Main Program --------

print("========== WELCOME TO THE BUDGET TRACKER ==========\n")

obj = BudgetTracker()     # Object Creation
obj.register()            # Calling register method
obj.login()               # Calling login method
```
