```python
def test8(a) :
  l = []
  for i in a :
    if type(i) == list:
      for j in i :
        l.append(j)
    else:
          if type(i) == int or type(i) == float:
            l.append(i)
  return l
```

```python
2num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

operator = input("Enter operator (+, -, *, /): ")

match operator:
    case "+":
        print("Result =", num1 + num2)

    case "-":
        print("Result =", num1 - num2)

    case "*":
        print("Result =", num1 * num2)

    case "/":
        if num2 != 0:
            print("Result =", num1 / num2)
        else:
            print("Error: Division by zero is not allowed.")

    case _:
        print("Invalid operator!")
```

```python
def test_fib (n):
  a,b = 0,1
  for i in range(n):
    yield a
    a,b = b,a+b
```

```python
for i in test_fib(10):
  print(i)
```
